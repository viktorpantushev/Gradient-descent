```python
import torch
import numpy as np
from IPython.display import display
import plotly.graph_objects as go
from ipywidgets import FloatSlider, interactive
```

```python
def f(x, y):
    return torch.sin(x) * torch.cos(y) * torch.exp(-(x**2 + y**2) / 5)

def plot_surface_plotly(xmin, xmax, ymin, ymax):
    x = torch.linspace(xmin, xmax, 30)
    y = torch.linspace(ymin, ymax, 30)
    X, Y = torch.meshgrid(x, y)
    Z = f(X, Y)

    # Create the surface plot
    fig = go.Figure(data=[go.Surface(z=Z.numpy(), x=X.numpy(), y=Y.numpy(), colorscale='Viridis', showscale=True)])

    dot_array_x = []
    dot_array_y = []

    for i in range(-10,10, 2):
        for j in range(-10,10, 2):
            dot_array_x.append(i)
            dot_array_y.append(j)



    # Generate some red points (you can customize their coordinates)
    red_x = torch.tensor(dot_array_x)
    red_y = torch.tensor(dot_array_y)
    red_z = f(red_x, red_y)  # Calculate Z values for red points

    # Add red points to the plot
    fig.add_trace(go.Scatter3d(
        x=red_x.numpy(),
        y=red_y.numpy(),
        z=red_z.numpy(),
        mode='markers',
        marker=dict(size=5, color='red', opacity=1),
        name='Red Points'
    ))

    fig.update_layout(title='Rotatable 3D Surface Plot with Red Points', scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ))
    fig.show()

# Define range sliders
xmin_slider = FloatSlider(value=-10, min=-10, max=0, step=0.1, description='X min:')
xmax_slider = FloatSlider(value=10, min=0, max=10, step=0.1, description='X max:')
ymin_slider = FloatSlider(value=-10, min=-10, max=0, step=0.1, description='Y min:')
ymax_slider = FloatSlider(value=10, min=0, max=10, step=0.1, description='Y max:')

# Create the interactive plot
interactive_plot = interactive(plot_surface_plotly, xmin=xmin_slider, xmax=xmax_slider, ymin=ymin_slider, ymax=ymax_slider)

# Display the widgets and interactive plot
interactive_plot
```
